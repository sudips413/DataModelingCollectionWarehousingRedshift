```python
import pandas as pd
```

```python
df= pd.read_csv('moviesv2.csv')
```

```python
# drop nan value of gross column

df.dropna(subset=['gross'], inplace=True)
df.dropna(subset=['runtime'], inplace=True)
df.dropna(subset=['no_of_reviews'], inplace=True)
df.dropna(subset=['rating'], inplace=True)
```

```python
df.to_csv('moviesv3.csv', index=False)
```

```python
df= pd.read_csv('moviesv3.csv')
df
```

```python
# join genre to single line removing \r\n

df['genres'] = df['genres'].str.replace('\r\n', '')

df['genres'] = df['genres'].str.replace('\t', '')
df
```
